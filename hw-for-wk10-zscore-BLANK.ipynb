Which albums stood out (AS IN THE BIGGEST)in terms of sales?

```python
import pandas as pd

from scipy.stats import zscore 
```

```python
lps = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv")
lps
```

```python
lps["lp_sales_zscore"] = zscore(lps["sales"])
lps

#round(lps.groupby("sales")["lp_sales_zscore"].agg([max]),0)
```

```python
lps_sales = round(lps.sort_values(by = ["sales"], ascending = False),0)
lps_sales
```

```python
lps_min = lps_sales["lp_sales_zscore"].agg(min)
lps_min
```

```python
lps_max = lps_sales["lp_sales_zscore"].agg(max)
lps_max

#lps_zscore = lps["lp_sales_zscore"].max()
#lps_zscore

#lps.value_counts("artist").to_frame("lp_sales_zscore").head(10)
```

```python
dfs = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv")
dfs
```

```python
dfs["stream_sales_zscore"] = zscore(dfs["streams_2023"])
dfs
```

```python
dfs_sales = round(dfs.sort_values(by = ["stream_sales_zscore"], ascending = False),0)
dfs_sales
```

```python
#preprocessing.normalize("stream_sales_zscore")
#dfs_sales.normalize(["stream_sales_zscore"])
```

```python
lp_stream_sales = pd.concat([lps_sales, dfs_sales], ignore_index=True)
lp_stream_sales
```

```python
lp_stream_sales['biggest'] =\
lp_stream_sales.apply(lambda x: 'LPS' if x['lp_sales_zscore'] > x['stream_sales_zscore'] else 'STREAM', axis=1)
lp_stream_sales.sample(20)
```

```python
lp_stream_sales['z-gap'] =\
lp_stream_sales.apply(lambda x: abs(x['lp_sales_zscore'] - x['stream_sales_zscore']), axis=1)
lp_stream_sales
```

```python
lp_stream_sales = lp_stream_sales.pivot(columns = "lp_sales_zscore",
        index = "album",
        values ="lp_sales_zscore")
lp_stream_sales
```
